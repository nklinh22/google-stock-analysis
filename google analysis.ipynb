```python
plt.figure(figsize=(12,6))
plt.plot(data['High'])
plt.title("Date vs Stock High Price",fontsize=30,color='r')
plt.xlabel("Date",fontsize=20,color='g')
plt.ylabel("Stock High Price",fontsize=20,color='g')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Volume'])
plt.title("Year wise volume",fontsize=30,color='r')
plt.xlabel("Year",fontsize=20,color='g')
plt.ylabel("Volume",fontsize=20,color='g')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Adj. Low'])
plt.title("Year wise Adj. Low",fontsize=30,color='r')
plt.xlabel("Year",fontsize=20,color='g')
plt.ylabel("Adj. Low",fontsize=20,color='g')
plt.show()
```

# Stock price trend over year

increase trend over years from 2012 to 2016

```python
plt.figure(figsize=(12,8))

glow = data['Adj. Low']
plt.plot(glow,color='g')
plt.plot(glow.rolling(window=2).mean(),color='r')
plt.plot(glow.rolling(window=2).std())
plt.xlabel('Year',fontsize=20,color='g')
plt.ylabel('Stock Price',fontsize=20,color='g')
```
